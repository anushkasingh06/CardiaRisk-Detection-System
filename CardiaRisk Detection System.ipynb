```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
df['Age'].plot(kind='hist', ax=axes[0, 0], title='Age')
df['Cholesterol'].plot(kind='hist', ax=axes[0, 1], title='Cholesterol')
df.head(20)['Blood Pressure'].value_counts().plot(kind='bar', ax=axes[1, 0], title='Blood Pressure')
df['Heart Rate'].plot(kind='hist', ax=axes[1, 1], title='Heart Rate')

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
df['Sex'].value_counts().plot(kind='bar', ax=axes[0, 0], title='Sex')
df['Diabetes'].value_counts().plot(kind='bar', ax=axes[0, 1], title='Diabetes')
df['Smoking'].value_counts().plot(kind='bar', ax=axes[1, 0], title='Smoking')
df['Previous Heart Problems'].value_counts().plot(kind='bar', ax=axes[1, 1], title='Previous Heart Problems')

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
axes[0].scatter(df['Age'].head(20), df['Blood Pressure'].head(20))
axes[0].set_xlabel('Age')
axes[0].set_ylabel('Blood Pressure')
axes[0].set_title('Age vs Blood Pressure')

axes[1].scatter(df['Cholesterol'].head(20), df['Heart Rate'].head(20))
axes[1].set_xlabel('Cholesterol')
axes[1].set_ylabel('Heart Rate')
axes[1].set_title('Cholesterol vs Heart Rate')

plt.tight_layout()
plt.show()
```

```python
Patient_ID = df.head(10).select_dtypes(include=['float64', 'int64'])

correlation_matrix = Patient_ID.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
df[['Age', 'Blood Pressure']].head(20).boxplot()
plt.title('Box Plot for Age and Blood Pressure')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.violinplot(data=df[['Cholesterol', 'Heart Rate']].head(20))
plt.title('Violin Plot for Cholesterol and Heart Rate')
plt.show()
```

```python
sns.pairplot(df[['Age', 'Cholesterol', 'Blood Pressure', 'Heart Rate']].head(20))
plt.suptitle('Pair Plot for Age, Cholesterol, Blood Pressure, and Heart Rate', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.plot(df['Exercise Hours Per Week'].head(20))
plt.title('Line Plot for Exercise Hours Per Week')
plt.xlabel('Index')
plt.ylabel('Exercise Hours')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
df['Stress Level'].head(20).plot(kind='area')
plt.title('Area Plot for Stress Level')
plt.xlabel('Index')
plt.ylabel('Stress Level')
plt.show()
```

```python
columns_for_pie = ['Sex', 'Diabetes', 'Smoking', 'Stress Level']

for column in columns_for_pie:
    value_counts = df[column].value_counts()
    plt.figure(figsize=(6, 6))
    plt.pie(value_counts, labels=value_counts.index, autopct='%1.1f%%', startangle=90)
    plt.title(f'{column} Distribution')
    plt.show()
```

```python
# Hypothesis 1: Cholesterol and Heart Attack
# Compare average cholesterol levels between individuals with and without heart attacks
cholesterol_heart_attack = df.groupby('Heart Attack Risk')['Cholesterol'].mean()
```

```python
# Visualization
sns.barplot(x=cholesterol_heart_attack.index, y=cholesterol_heart_attack.values)
plt.title('Average Cholesterol Levels by Heart Attack Risk')
plt.xlabel('Heart Attack Risk')
plt.ylabel('Average Cholesterol Level')
plt.show()
```

```python
# Hypothesis 3: Lifestyle factors and Heart Attack
# Compare the distribution of smoking, obesity, and exercise among individuals with and without heart attacks
sns.countplot(x='Smoking', hue='Heart Attack Risk', data=df)
plt.title('Smoking Distribution by Heart Attack Risk')
plt.show()

sns.countplot(x='Obesity', hue='Heart Attack Risk', data=df)
plt.title('Obesity Distribution by Heart Attack Risk')
plt.show()

sns.boxplot(x='Heart Attack Risk', y='Exercise Hours Per Week', data=df)
plt.title('Exercise Hours Per Week by Heart Attack Risk')
plt.show()
```

```python
# Hypothesis 4: Age and Heart Attack
# Compare the average age of individuals with and without heart attacks
age_heart_attack = df.groupby('Heart Attack Risk')['Age'].mean()

# Visualization
sns.barplot(x=age_heart_attack.index, y=age_heart_attack.values)
plt.title('Average Age by Heart Attack Risk')
plt.xlabel('Heart Attack Risk')
plt.ylabel('Average Age')
plt.show()
```

```python
# Hypothesis 5: Stress Level and Sedentary Hours
# Explore the relationship between stress level, sedentary hours, and heart attack risk
sns.boxplot(x='Heart Attack Risk', y='Stress Level', data=df)
plt.title('Stress Level by Heart Attack Risk')
plt.show()

sns.boxplot(x='Heart Attack Risk', y='Sedentary Hours Per Day', data=df)
plt.title('Sedentary Hours Per Day by Heart Attack Risk')
plt.show()
```
